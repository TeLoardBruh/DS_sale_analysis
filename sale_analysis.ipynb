```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from itertools import combinations
from collections import Counter
```

## Merging all 12 months of the sales into 1 files

```python
# df = pd.read_csv('./Sales_Data/Sales_April_2019.csv')
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])


# moving it to one file
all_months_data.to_csv('all_months_data.csv',index=False)
```

## Start to use the new file

```python
all_data = pd.read_csv('./all_months_data.csv')
all_data.head()
```

## Start analysis

### Data cleaning

#### Drop NAN rows

```python
# checking for NAN
nan_data = all_data[all_data.isna().any(axis=1)]
nan_data.head()

# dropping NAN
all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'Or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

#### Converting Cols to correct types

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns

#### Add month columns

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Add sales columns

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Add city column

```python
# adding city columns 
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

### Question 1 : what was the best month for sales ? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
results
```

#### Drawing a graph to show the 12 months of sales

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in in USD ($)')
plt.xlabel('Month number')
plt.show()
```

### Question 2 :  Which city have highest numbers of sales?

```python
results_city = all_data.groupby('City').sum()
results_city
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results_city['Sales'])
plt.xticks(cities, rotation='vertical',size=8)
plt.ylabel('Sales in each city')
plt.xlabel('City')
plt.show()
```

### Question 3 : What time should we display the product advertisement to maximize the likelihood of customer's buying a product ?

```python
all_data.head()
```

```python
# converting order date to datetime object 

all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
# add hours column
all_data['Hour'] = all_data['Order Date'].dt.hour

# add minute column
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data['Count'] = 1
all_data.head()
```

```python
# grouping the data and ploting

hours = [hour for hour, df in all_data.groupby('Hour')]

# plotting graph
xpoints = np.array(hours)
ypoints = np.array(all_data.groupby(['Hour']).count())

plt.plot(xpoints, ypoints)
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

### Question 4 What product are most often sold together?

```python
# grouping data with the same order id

df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped Product'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))


# dropping the duplicated line since merging
df = df[['Order ID', 'Grouped Product']].drop_duplicates()

df.head()
```

```python
# start to count pairs of the group product

count = Counter()

for row in df['Grouped Product']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    

for key, value in count.most_common(20):
    print(key, value)
```

### Question 5 : What product have sold the most?

```python
all_data.head()
```

```python
product_grouped = all_data.groupby('Product')
quantity_ordered = product_grouped.sum()['Quantity Ordered']

products = [product for product, df in product_grouped]

plt.bar(products, quantity_ordered)
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(products, rotation='vertical',size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
